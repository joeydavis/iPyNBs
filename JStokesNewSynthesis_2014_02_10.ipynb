```python
def getSlopes(l, t=[240, 480, 960]):
    slopeDict = {}
    for p in l.keys():
        v = l[p]
        slopeDict[p] = max(numpy.polyfit(t, v, 1)[0],0.0)
    return slopeDict
```

```python
slopesDMSO = getSlopes(allListDMSO)
slopesL3 = getSlopes(allListL3)
slopesC3 = getSlopes(allListC3)
slopesL6 = getSlopes(allListL6)
slopesC6 = getSlopes(allListC6)
```
