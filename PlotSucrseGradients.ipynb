```python
import qMS
import vizLib
import qMSDefs
import pandas
import numpy
import matplotlib.pyplot as plt
import pylab
import glob
```

```python
path = '/home/jhdavis/McMasterMSUDataSets/PulseLabeling/2013_08_08-MSUPulse3/handFilt/'
```

```python
files = glob.glob(path+'*.tsv')
```

```python
vizLib.setRcs(scale=12, legendScale=16)
```

```python
t5_10 = pandas.read_table(files[0],names=['time', 'A260'], index_col=False)
t5_1000 = pandas.read_table(files[1],names=['time', 'A260'], index_col=False)

t10_70S = t5_10[76:270]
t10_45S = t5_10[295:425]

t1000_70S = t5_1000[45:265]
t1000_50S = t5_1000[267:350]

t10_offset = -.125
t1000_offset = .1

t10_70SArea = (t10_70S['A260']+t10_offset).sum(axis=1)/38
t10_45SArea = (t10_45S['A260']+t10_offset).sum(axis=1)/25
t10_70SAreaN = t10_70SArea/t10_70SArea
t10_45SAreaN = t10_45SArea/t10_70SArea

t1000_70SArea = (t1000_70S['A260']+t1000_offset).sum(axis=1)/38
t1000_50SArea = (t1000_50S['A260']+t1000_offset).sum(axis=1)/25
t1000_70SAreaN = t1000_70SArea/t1000_70SArea
t1000_50SAreaN = t1000_50SArea/t1000_70SArea
```

```python
f = pylab.figure(figsize=(6,7.2))
ax = f.add_subplot(111)
ax.plot(t5_10['time'],t5_10['A260']+t10_offset, label='10 $\mu$M IPTG\n45S:70S = '+str(round(t10_45SAreaN,1)), lw=3, color='#e31a1c')
ax.plot(t5_1000['time'],t5_1000['A260']+t1000_offset, label='1 mM IPTG\n50S:70S = '+str(round(t1000_50SAreaN,1)), lw=3, color='#377db8')
ax.set_xlim((2.5, 11.4))
ax.set_ylim((0, 1))
ax.set_xticks([])
ax.set_yticks([])
ax.fill(t10_70S['time'], t10_70S['A260']+t10_offset, facecolor='#e31a1c', alpha=0.35, edgecolor='none')
ax.fill(t10_45S['time'], t10_45S['A260']+t10_offset, facecolor='#e31a1c', alpha=0.65, edgecolor='none')
ax.fill(t1000_70S['time'], t1000_70S['A260']+t1000_offset, facecolor='#377db8', alpha=0.35, edgecolor='none')
ax.fill(t1000_50S['time'], t1000_50S['A260']+t1000_offset, facecolor='#377db8', alpha=0.65, edgecolor='none')

x70S = 4.8
ax.plot([x70S, x70S], [0, 0.875], color='grey', linestyle='--', lw=2)
ax.text(x70S, 0.89, '70S', size=18, weight=500, horizontalalignment='center')

x30S = 10.25
ax.plot([x30S, x30S], [0, 0.3], color='grey', linestyle='--', lw=2)
ax.text(x30S, 0.315, '30S', size=18, weight=500, horizontalalignment='center')

x50S = 7.25
ax.plot([x50S, x50S], [0, 0.35], color='grey', linestyle='--', lw=2)
ax.text(x50S-0.2, 0.365, '50S', size=18, weight=500, horizontalalignment='center')

x45S =7.8
ax.plot([x45S, x45S], [0, 0.35], color='grey', linestyle='--', lw=2)
ax.text(x45S+.2, 0.365, '45S', size=18, weight=500, horizontalalignment='center')


ax.text(4.65, 0.1, str(round(t10_70SAreaN,2)), size=16, weight=500)
ax.text(7.65, 0.1, str(round(t10_45SAreaN,2)), size=16, weight=500)
ax.text(4.5, 0.4, str(round(t1000_70SAreaN,2)), size=16, weight=500)
ax.text(7.15, 0.33, str(round(t1000_50SAreaN,2)), size=16, weight=500)

pylab.legend()
pylab.tight_layout()
#pylab.savefig('/home/jhdavis/Dropbox/fromLinux/gradientProfiles.png')
#pylab.savefig('/home/jhdavis/Dropbox/fromLinux/gradientProfiles.pdf')
```
