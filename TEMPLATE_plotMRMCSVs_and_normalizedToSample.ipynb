```python
import vizLib
import qMS
import mrmTools
import pylab
import matplotlib
import numpy
import scipy
import pandas
import glob
```

```python
path = '/home/jhdavis/data/2015_03_09-BTsMRM/'
figPath = '/home/jhdavis/data/2015_03_09-BTsMRM/'
saveFigs = True
figCounter=3
```

```python
fileName = glob.glob(path+'/*.csv')[3]
fileName
```

```python
allPeps = mrmTools.readMRMCSV(fileName, fileNameHeader = 'File Name')
goodPeps = allPeps[allPeps['allClear'] == True]
```

```python
samples = qMS.sort_nicely(list(allPeps['File Name'].unique()))
labels = ['wt_30S', 'rimM_30S', 'yjeQ_30S', 'wt_70S', 'rimM_30S_2x', 'yjeQ_30S_2x']
labelHash = {samples[i]:labels[i] for i in range(len(samples))}
colors = vizLib.getBCs('q', 6)
```

```python
yMax=7.25
orderedProteins = ['Era', 'RbfA', 'RimM', 'yjeQ', 'L2', 'L20', 'L24'] + ['S'+str(i) for i in range(1,22)]
vizLib.setRcs(scale=16, legendScale=18, axisLineWidth=1, tickScale=1)
ax = vizLib.plotMRMCsv(goodPeps, samples=samples, labelHash=labelHash, yAxisLabel='14N/15N', legendCols=3, legendLoc='upper left', legendBBox=(0, 1.0), 
                  proteins = orderedProteins, alpha=0.5, grid=True, yTicks=3, medianMarkerInc=1.25, yAxis = [0, 1, 2, 3, 4, 5, 6, 7],
                  colors=None, figSize = (20,5), markersize=10, yMax=yMax)
ax.set_ylim(-0.5, yMax)
pylab.tight_layout()
if saveFigs:
    pylab.savefig(figPath+str(figCounter)+'-filteredExpSamplesAllProteins.png')
    pylab.savefig(figPath+str(figCounter)+'-filteredExpSamplesAllProteins.pdf')
    figCounter=figCounter+1
```

```python
yMax=3.5
orderedProteins = ['Era', 'RbfA', 'RimM', 'yjeQ', 'L2', 'L20', 'L24'] + ['S'+str(i) for i in range(1,22)]
ax = vizLib.plotMRMCsv(goodPeps, samples=samples, labelHash=labelHash, yAxisLabel='14N/15N [norm S4]', 
                       legendCols=3, legendLoc='upper right', legendBBox=(0.5, 1.0), proteins = orderedProteins, 
                       alpha=0.5, grid=True, yTicks=3, medianMarkerInc=1.25, yAxis = [0, 1, 2, 3], normSample=None,
                      colors=None, figSize = (20,5), markersize=10, yMax=yMax, normProtein = 'S4')
ax.set_ylim(-0.25, yMax)
pylab.tight_layout()
if saveFigs:
    pylab.savefig(figPath+str(figCounter)+'-filteredExpSamplesAllProteins_normS4.png')
    pylab.savefig(figPath+str(figCounter)+'-filteredExpSamplesAllProteins_normS4.pdf')
    figCounter=figCounter+1
```
