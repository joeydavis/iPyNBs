```python
colors = vizLib.getBCs('q', 5)
vizLib.setRcs(axisLineWidth=1)
columns = ['sMRM_coli70S_6'+str(i)+' Area' for i in range(2,7)]
f = pylab.figure(figsize=(22,16))
cList = ['Blues', 'Reds', 'Greens', 'Oranges', 'Purples']
for i,n in enumerate(columns):
    a = f.add_subplot(5,4,4*i+1)
    colors = vizLib.getBCs('s', 6, cList[i])
    selection = coli[coli['IsotopeLabelType']=='light']
    toPlot = selection[[n,'sMRM_coli70S_61 Area']]
    toPlot = toPlot.dropna()
    xs = range(len(toPlot[n]))
    a.plot(xs, toPlot[n]/toPlot['sMRM_coli70S_61 Area'], 'o', color='none', markersize=8, markeredgecolor=colors[4], markeredgewidth=1, label='light_5^'+str(i+1))
    #a.plot(xs, [0.04]*len(xs), '--', color='grey')
    a.plot(xs, [1.0/(6**(i+1))]*len(xs), '--', color='grey')
    #a.plot(xs, [1]*len(xs), '--', color='grey')
    #a.plot(xs, [5]*len(xs), '--', color='grey')
    #a.plot(xs, [25]*len(xs), '--', color='grey')
    a.set_yscale('log')
    a.set_xticks([])
    a.set_ylabel('expDilution / undilted')
    a.set_title('light')
    vizLib.cleanAxis(a)
    
    
    a = f.add_subplot(5,4,4*i+2)
    selection = coli[coli['IsotopeLabelType']=='heavy']
    toPlot = selection[[n,'sMRM_coli70S_61 Area']]
    toPlot = toPlot.dropna()
    xs = range(len(toPlot[n]))
    a.plot(xs, toPlot[n]/toPlot['sMRM_coli70S_61 Area'], 'o', color='none', markersize=8, markeredgecolor=colors[4], markeredgewidth=1, label='heavy_5^'+str(i+1))
    a.plot(xs, [1.0/(6**(i+1))]*len(xs), '--', color='grey')
    #a.plot(xs, [0.2]*len(xs), '--', color='grey')
    #a.plot(xs, [1]*len(xs), '--', color='grey')
    #a.plot(xs, [5]*len(xs), '--', color='grey')
    #a.plot(xs, [25]*len(xs), '--', color='grey')
    a.set_yscale('log')
    a.set_xticks([])
    a.set_title('heavy')
    vizLib.cleanAxis(a)
    
    a = f.add_subplot(5,4,4*i+3)
    selection = coli[coli['IsotopeLabelType']=='heavy']
    toPlot = selection[['sMRM_coli70S_6'+str(i+2)+' TotalArea','sMRM_coli70S_61 TotalArea']]
    toPlot = toPlot.dropna()
    toPlot = toPlot.drop_duplicates()
    xs = range(len(toPlot['sMRM_coli70S_6'+str(i+2)+' TotalArea']))
    a.plot(xs, toPlot['sMRM_coli70S_6'+str(i+2)+' TotalArea']/toPlot['sMRM_coli70S_61 TotalArea'], 'o', color='none', markersize=8, markeredgecolor=colors[4], markeredgewidth=1, label='heavy_5^'+str(i+1))
    #a.plot(xs, [0.04]*len(xs), '--', color='grey')
    #a.plot(xs, [0.2]*len(xs), '--', color='grey')
    a.plot(xs, [1.0/(6**(i+1))]*len(xs), '--', color='grey')
    #a.plot(xs, [1]*len(xs), '--', color='grey')
    #a.plot(xs, [5]*len(xs), '--', color='grey')
    #a.plot(xs, [25]*len(xs), '--', color='grey')
    a.set_yscale('log')
    a.set_xticks([])
    a.set_title('light_total')
    vizLib.cleanAxis(a)
    
    a = f.add_subplot(5,4,4*i+4)
    selection = coli[coli['IsotopeLabelType']=='heavy']
    toPlot = selection[['sMRM_coli70S_6'+str(i+2)+' TotalArea','sMRM_coli70S_61 TotalArea']]
    toPlot = toPlot.dropna()
    toPlot = toPlot.drop_duplicates()
    xs = range(len(toPlot['sMRM_coli70S_6'+str(i+2)+' TotalArea']))
    a.plot(xs, toPlot['sMRM_coli70S_6'+str(i+2)+' TotalArea']/toPlot['sMRM_coli70S_61 TotalArea'], 'o', color='none', markersize=8, markeredgecolor=colors[4], markeredgewidth=1, label='heavy_5^'+str(i+1))
    #a.plot(xs, [0.04]*len(xs), '--', color='grey')
    #a.plot(xs, [0.2]*len(xs), '--', color='grey')
    a.plot(xs, [1.0/(6**(i+1))]*len(xs), '--', color='grey')
    #a.plot(xs, [1]*len(xs), '--', color='grey')
    #a.plot(xs, [5]*len(xs), '--', color='grey')
    #a.plot(xs, [25]*len(xs), '--', color='grey')
    a.set_yscale('log')
    a.set_xticks([])
    a.set_title('heavy_total')
    vizLib.cleanAxis(a)

    
    
    #pylab.legend(loc='lower right')
    pylab.savefig(path+'individualAreas.png', dpi=300, bbox_inches='tight')
```
