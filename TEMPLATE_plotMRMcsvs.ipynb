```python
import vizLib
import qMS
import mrmTools
import pylab
import matplotlib
import numpy
import scipy
import pandas
import glob
```

```python
path = '/home/jhdavis/data/2015_01_12-GB30Sanalysis'
figPath = '/home/jhdavis/data/2015_01_12-GB30Sanalysis/'
saveFigs = False
vizLib.setRcs(axisLineWidth=1)
```

```python
fileName = glob.glob(path+'/*.csv')[1]
fileName
```

```python
allFiles = mrmTools.readMRMCSV(fileName, fileNameHeader = 'Replicate Name')
goodPeps = allFiles[allFiles['allClear'] == True]
```

```python
factors = ['L20', 'Era', 'RbfA', 'RimM']
small = ['S'+str(i) for i in range(1,20)]
small = small + ['S20L26', 'S21']
f_and_s = factors+small
```

```python
GBSet2Names = ['GB_finalsMRM_2-'+str(i) for i in range(1,8)]
GB2controls = GBSet2Names[0:4]
GB2experimental = GBSet2Names[4:]
```

```python
con = mrmTools.getAllOccupancyFileListMRM(goodPeps, GB2controls, num = ['light'], den = ['heavy'], selectField='Replicate Name')
conNorm = mrmTools.getAllOccupancyFileListMRM(goodPeps, GB2controls, num = ['light'], den = ['heavy'], 
                                              selectField='Replicate Name', normProtein = 'S4')
```

```python
colors = vizLib.getBCs('s', 4, name='reds')
nameList = [(i,i) for i in GB2controls]
```

```python
a = vizLib.makePlotWithStatsDictDict(conNorm, figSize = (15, 5), legend=True, legendCols=2, yMax=4, title='factor titration', alpha = .25,
                                 colors=colors,  highlightMed=True, namesList=nameList, subunits = f_and_s, yAxis = '14N/15N [normS4]')
a.set_yticks([0, 1, 2, 3, 4])
vizLib.cleanAxis(a, ticksOnly=True)
pylab.savefig(figPath+'factorTitration_GB2_normS4.png')
pylab.savefig(figPath+'factorTitration_GB2_normS4.pdf')
```

```python
a = vizLib.makePlotWithStatsDictDict(con, figSize = (15, 5), legend=True, legendCols=2, yMax=4, title='factor titration', alpha = .25,
                                 colors=colors,  highlightMed=True, namesList=nameList, subunits = f_and_s, yAxis = '14N/15N')
a.set_yticks([0, 1, 2, 3, 4])
vizLib.cleanAxis(a, ticksOnly=True)
pylab.savefig(figPath+'factorTitration_GB2.png')
pylab.savefig(figPath+'factorTitration_GB2.pdf')
```

```python
exp = mrmTools.getAllOccupancyFileListMRM(goodPeps, GB2experimental, num = ['light'], den = ['heavy'], selectField='Replicate Name')
expNorm = mrmTools.getAllOccupancyFileListMRM(goodPeps, GB2experimental, num = ['light'], den = ['heavy'], 
                                              selectField='Replicate Name', normProtein = 'S4')
```

```python
colors = vizLib.getBCs('q', 3)
nameList = [['GB_finalsMRM_2-5', 'GB2-5_WT30S'], ['GB_finalsMRM_2-6', 'GB2-6_dYjeQ'], ['GB_finalsMRM_2-7', 'GB2-7_dRimM']]
```

```python
a = vizLib.makePlotWithStatsDictDict(expNorm, figSize = (15, 5), legend=True, legendCols=3, yMax=2.5, title='experimental', alpha = .25,
                                 colors=colors,  highlightMed=True, namesList=nameList, subunits = f_and_s, yAxis = '14N/15N [normS4]')
a.set_yticks([0, 1, 2])
vizLib.cleanAxis(a, ticksOnly=True)
pylab.savefig(figPath+'experimental_GB2_normS4.png')
pylab.savefig(figPath+'experimental_GB2_normS4.pdf')
```

```python
a = vizLib.makePlotWithStatsDictDict(exp, figSize = (15, 5), legend=True, legendCols=3, yMax=3, title='experimental', alpha = .25,
                                 colors=colors,  highlightMed=True, namesList=nameList, subunits = f_and_s, yAxis = '14N/15N')
a.set_yticks([0, 1, 2, 3])
vizLib.cleanAxis(a, ticksOnly=True)
pylab.savefig(figPath+'experimental_GB2.png')
pylab.savefig(figPath+'experimental_GB2.pdf')
```
