```python
import qMS
import pylab
import glob
import pandas
import numpy
import vizLib
import qMSDefs
```

```python
path = '/home/jhdavis/data/2013_05_28-MSUPulse2/13to20'
files = glob.glob(path+'/*_filt_hand.csv')
files.sort()
num = ['AMP_L', 'AMP_U']
den = ['AMP_U', 'AMP_L', 'AMP_S']
offset=0.0
markerSize=12
color='#e31a1c'
yMax = 1.5
vizLib.setRcs(scale=12, legendScale=10)
s50S = qMSDefs.subtilisLargeSubunit
cs = qMSDefs.diverging
figSize = (15, 6)
```

```python
vizLib.makePlotWithFileList(files, num, den, AllProteins=s50S, normProtein='BSubL24', yMax=1.2, colors=cs, median=True, figSize=figSize)
```

```python
allStats = qMS.multiStatsDict(files, num, den, normalization=1.0, offset=0.0, normProtein='BSubL24')
```

```python
depMeds = []
zeroProts = ['BSubL28','BSubL33a','BSubL35','BSubL36']
for k in zeroProts:
    for i in allStats.keys():
        try:
            if not (str(numpy.median(allStats[i][k])) == 'nan'):
                depMeds.append(numpy.median(allStats[i][k]))
            else:
                print "nan"
        except KeyError:
            print 'key error with ' + k
                                  
    
```
