```python
def calcCoverage(L):
    totalChars = len(L[0])
    finalString = list(L[0])
    for i in L:
        addString = list(L)
        for j in range(0,len(i)):
            if not (i[j] == '-'):
                finalString[j] = i[j]
    fs = "".join(finalString)
    cov = round(1.0 - float(fs.count('-'))/float(len(fs)),2)
    return cov
        
    
```

```python
outfile = open('L2CoverageTTOF.txt', 'w')
protHash = {}
for i in proteins:
    stringL = []
    for j in range(1,5):
        stringL.append(list('-'*(len(gd[i]['AA'])-1)))
        srs = df[df['protein'] == i][df[df['protein'] == i]['charge']==j]['startres'].values
        seqs = df[df['protein'] == i][df[df['protein'] == i]['charge']==j]['seq'].values
        for n,k in enumerate(srs):
            stringL[j-1][int(k-1):int(k-1+len(seqs[n]))] = seqs[n]
        string = "".join(stringL[j-1])
        cov = round(1.0 - float(string.count('-'))/float(len(string)),2)
        outfile.write(str(cov) + ":\t" + string + '\t+' + str(j) + '\n')
    try:
        protHash[i] = calcCoverage(stringL)
        calcCoverage(stringL)
    except IndexError:
        print list(gd[i]['AA'])
        print stringL[0]
        print stringL[1]
    outfile.write(i + ':\t' + gd[i]['AA'][:-1] + '\t' + str(protHash[i])+'\n\n')
outfile.flush()
outfile.close()
```
