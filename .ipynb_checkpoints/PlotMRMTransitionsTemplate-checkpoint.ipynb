```python
import qMS
import vizLib
import glob
import pylab
import pandas
import numpy
import qMSDefs
import mrmTools
import matplotlib.pyplot as plt
from matplotlib import gridspec
```

```python
vizLib.css_styling(half=False)
```

```python
vizLib.setRcs(axisLineWidth=1, tickScale=0.5, legendScale=12)
path = '/home/jhdavis/data/2014/2014_04_28-BrittonPLOSMRM/'
files = glob.glob(path+'*.tsv')
dataFrame = pandas.read_csv(files[0], na_values='#N/A', sep='\t')
plos20T = dataFrame[dataFrame['FileName'] =='plos20T.wiff']
```

```python
biguns = plos20T[(plos20T['TotalArea'] > 1500000)]
biguns = biguns.drop_duplicates(cols='PeptideModifiedSequence')
toPlotIndex = biguns.index.values
toPlotIndex = list(toPlotIndex)
toPlotIndex.remove(3288)
toPlotIndex.remove(3328)
toPlotIndex = numpy.array(toPlotIndex)
```

```python
fig = pylab.figure(figsize=(25,15))
axisArray = []
tickNum=2
for i in range(16):
    a = fig.add_subplot(4,4,i+1)
    a = mrmTools.plotTotalChromPairsMRM(dataFrame, toPlotIndex[i], a, smooth=5, zoom=True)
    info = mrmTools.getInfoMRM(dataFrame, toPlotIndex[i])
    #a.set_title(toPlotIndex[i])
    a.set_title(info[1])
    vizLib.tickNum(a, xAxis=3, yAxis=5)
    vizLib.cleanAxis(a, ticksOnly=True)
    a.legend()
fig.tight_layout()
#pylab.savefig('/home/jhdavis/Dropbox/allTots.png', dpi=300, bbox_inches='tight')
```

```python
fig = pylab.figure(figsize=(25,15))
axisArray = []
for i in range(16):
    a = fig.add_subplot(4,4,i+1)
    a = mrmTools.plotAllTransitionsMRM(dataFrame, toPlotIndex[i], a, smooth=5, zoom=True)
    info = mrmTools.getInfoMRM(dataFrame, toPlotIndex[i])
    a.set_title(info[1])
    #a.set_title(toPlotIndex[i])
    vizLib.tickNum(a, xAxis=4, yAxis=4)
    vizLib.cleanAxis(a, ticksOnly=True)
fig.tight_layout()
#pylab.savefig('/home/jhdavis/Dropbox/trans.png', dpi=300, bbox_inches='tight')
```
