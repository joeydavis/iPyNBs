```python
f = pylab.figure(figsize=(22, 33))
def doPlot(i,j,ax, allValues):
    ax.scatter(i, j)
    for prot, xl, yl in zip(allValues, i, j):
        ax.annotate(str(prot), xy = (float(xl), float(yl)), xytext = (15,15), textcoords = 'offset points', arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
    m,b = numpy.polyfit(i, j,1)
    xs = numpy.linspace(min(i.min(), j.min()), max(i.max(), j.max()),200)
    ax.plot(xs, xs*m+b)
    corr = stats.pearsonr(i, j)
    ax.text(0.5, 0.15, 'pearson correlation = '+str(round(float(corr[0]),2)), size=20, transform=ax.transAxes)
    return ax

a = f.add_subplot(321)
i = everything['Synthesis']
j = everything['cm1']
a = doPlot(i,j,a,everything.index.values)
a.set_ylabel('Cm 1 hour [14N/15N] norm L5')
a.set_xlabel('Siibak synthesis')

b = f.add_subplot(322)
i = everything['Synthesis']
j = everything['cm6']
b = doPlot(i,j,b,everything.index.values)
b.set_ylabel('Cm 6 hour [14N/15N] norm L5')
b.set_xlabel('Siibak synthesis')

c = f.add_subplot(323)
i = everything['Synthesis']
j = everything['lam1']
c = doPlot(i,j,c,everything.index.values)
c.set_ylabel('Lam 1 hour [14N/15N] norm L5')
c.set_xlabel('Siibak synthesis')

d = f.add_subplot(324)
i = everything['Synthesis']
j = everything['lam6']
d = doPlot(i,j,d,everything.index.values)
d.set_ylabel('Lam 6 hour [14N/15N] norm L5')
d.set_xlabel('Siibak synthesis')
pylab.tight_layout()

c = f.add_subplot(325)
i = everything['Synthesis']
j = everything['dmso1']
c = doPlot(i,j,c,everything.index.values)
c.set_ylabel('DMSO 1 hour [14N/15N] norm L5')
c.set_xlabel('Siibak synthesis')

d = f.add_subplot(326)
i = everything['Synthesis']
j = everything['dmso6']
d = doPlot(i,j,d,everything.index.values)
d.set_ylabel('DMSO 6 hour [14N/15N] norm L5')
d.set_xlabel('Siibak synthesis')
pylab.tight_layout()

pylab.savefig(figPath+'pearson0.png')
```
