```python
rawDataWTS = qMS.readDataFile(figPath+'toClusterNormL5Ratio.txt', 1.0, delimiter='\t')
rawMapWTS = vizLib.drawHeatMap(rawDataWTS, "", nameList=labels, colors=pylab.cm.Reds, scale='14N/[15N]', saveName=figPath+'unclusteredHeatMap.png', figSize=(5.5,8.5), colorBar=True)
```

```python
cDataWTS = clusteringModule.clusterData(rawDataWTS, rowMethod=True, columnMethod=True, method='ward', metric='euclidean')
cOrder=[i-1 for i in [4,6,1,3,2,5]]
cMapWTS = vizLib.drawHeatMap(cDataWTS, "", topDendro=False, dendro=True, nameList=[labels[i] for i in cOrder], saveName=figPath+'clusteredHeatMap.png', figSize=(5.5,8.5), colorBar=False)
```

```python
siibakDataL = pandas.read_csv(figPath+'SiibakLarge.txt', sep='\t', index_col='Protein')
siibakDataS = pandas.read_csv(figPath+'SiibakSmall.txt', sep='\t', index_col='Protein')
siibakDataL = siibakDataL.fillna(0.0)
siibakDataS = siibakDataS.fillna(0.0)
siibakDataAll = siibakDataS.append(siibakDataL)
```

```python
myData = pandas.read_csv(figPath+'toScatter.txt', sep='\t', index_col='Protein')
everything = myData.join(siibakDataAll)
```

```python
f = pylab.figure(figsize=(22, 17))
a = f.add_subplot(221)

a.scatter(everything['Synthesis'], everything['cm1'])
for prot, xl, yl in zip(everything.index.values, everything['Synthesis'].values, everything['cm1'].values):
        a.annotate(str(prot), xy = (float(xl), float(yl)), xytext = (15,15), textcoords = 'offset points', arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
a.set_ylabel('Cm 1 hour [14N/15N] norm L5')

b = f.add_subplot(222)
b.scatter(everything['Synthesis'], everything['cm6'])
for prot, xl, yl in zip(everything.index.values, everything['Synthesis'].values, everything['cm6'].values):
        b.annotate(str(prot), xy = (float(xl), float(yl)), xytext = (15,15), textcoords = 'offset points', arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
b.set_ylabel('Cm 6 hour [14N/15N] norm L5')

c = f.add_subplot(223)
c.scatter(everything['Synthesis'], everything['lam1'])
for prot, xl, yl in zip(everything.index.values, everything['Synthesis'].values, everything['lam1'].values):
        c.annotate(str(prot), xy = (float(xl), float(yl)), xytext = (15,15), textcoords = 'offset points', arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
c.set_ylabel('Lam 1 hour [14N/15N] norm L5')
c.set_xlabel('siibak synthesis')

d = f.add_subplot(224)
d.scatter(everything['Synthesis'], everything['lam6'])
for prot, xl, yl in zip(everything.index.values, everything['Synthesis'].values, everything['lam6'].values):
        d.annotate(str(prot), xy = (float(xl), float(yl)), xytext = (15,15), textcoords = 'offset points', arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
d.set_ylabel('Lam 6 hour [14N/15N] norm L5')
d.set_xlabel('siibak synthesis')
pylab.tight_layout()
pylab.savefig(figPath+'correlationSiibak.png')
```
