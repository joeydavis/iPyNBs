```python
setName = 'W4'
saveFigs=False
vizLib.setRcs(axisLineWidth=1, legendScale=12)
f = pylab.figure(figsize = (11, 8.5))
a = f.add_subplot(121)
xs = range(1, len(primerNames)+1)
ys = [namedHash[i][2] for i in primerNames]
a.bar(xs, ys, width=0.4, color='black', label='mapped reads')
xs = [i+0.4 for i in xs]
ys = [namedHash[i][3][0] + namedHash[i][3][1] for i in primerNames]
a.bar(xs, ys, width=0.4, color='lightgrey', label='bc assigned reads')
a.legend()
a.set_xticks([i for i in xs])
a.set_xticklabels(primerNames, rotation=90)
a.set_xlim(0, len(primerNames)+2)
a.set_ylabel('number of reads')
a.set_xlabel('primerID')
a.set_title(setName+'_mappedReads_ham1_rl14')
vizLib.cleanAxis(a)

b = f.add_subplot(122)
xs = range(1, len(primerNames)+1)
ysUpper = [namedHash[i][3][0] for i in primerNames]
ysLower = [0-namedHash[i][3][1] for i in primerNames]
b.bar(xs, ysUpper, width=0.8, color='black', label='RRRY reads')
b.bar(xs, ysLower, width=0.8, color='lightgrey', label='YYYR reads')
b.legend()
b.set_xticks([i+0.4 for i in xs])
b.set_xticklabels(primerNames, rotation=90)
b.set_xlim(0, len(primerNames)+2)
b.set_ylabel('number of reads')
b.set_xlabel('primerID')
b.set_title(setName+'_mappedReads_ham1_rl14')
vizLib.cleanAxis(b)
pylab.tight_layout()
if saveFigs:
    pylab.savefig(figPath+str(figCounter)+setName+'mappedReads.png', dpi=300)
    figCounter +=1
```

# Plot fraction of reads for each primer for all datasets

```python
readsToUse = 100000
rootPath = '/home/jhdavis/data/2014/2014_06_07-ShapeSeqAnalysis/data/'
filesToUse = [[rootPath+'W3/W3_TTAGGC_L001_R1_001.fastq', rootPath+'W3/W3_TTAGGC_L001_R2_001.fastq'],
              [rootPath+'W4/W4_TGACCA_L001_R1_001.fastq', rootPath+'W4/W4_TGACCA_L001_R2_001.fastq'],
              [rootPath+'W22/W22_CGTACG_L001_R1_001.fastq', rootPath+'W22/W22_CGTACG_L001_R2_001.fastq'],
              [rootPath+'W23/W23_GAGTGG_L001_R1_001.fastq', rootPath+'W23/W23_GAGTGG_L001_R2_001.fastq'],
              [rootPath+'W24/W24_GGTAGC_L001_R1_001.fastq', rootPath+'W24/W24_GGTAGC_L001_R2_001.fastq']]
```

```python
assHat = {}
for readPair in filesToUse:
    fwdReadFile = readPair[0]
    revReadFile = readPair[1]
    name = fwdReadFile.split('/')[-1]
    assHat[name] = [NGSTools.assignReadsToPrimers(primerList, fwdReadFile, revReadFile, readsToUse=readsToUse,
                    maxHamDist=1, rtMatch=13, adapterLength=30, outputPath=None, trunc=35, writeFile=False)]
    namedHash = {}
    for k in assHat[name][0].keys():
        if not k == 'noMatch':
            namePL = primerList[primerList['rtPrimer'].str.slice(0,13) == k]['name'].values[0]
        else:
            namePL = 'noMatch'
        namedHash[namePL] = assHat[name][0][k][2]
    assHat[name].append(namedHash)
```

```python
primerNames = list(primerList['name'].values)
primerNames.append('noMatch')
```

```python
vizLib.setRcs(axisLineWidth=1, legendScale=12)
f = pylab.figure(figsize = (11, 8.5))
a = f.add_subplot(111)
xs = range(1, len(primerNames)+1)
bcFileNames = assHat.keys()
bcFileNames = qMS.sort_nicely(bcFileNames)
colors = vizLib.getBCs('q', 5)
for iterator, barcode in enumerate(bcFileNames):
    ys = [assHat[barcode][1][i] for i in primerNames]
    a.bar(xs, ys, width=0.15, color=colors[iterator], label=barcode)
    xs = [i+0.15 for i in xs]

a.legend()
a.set_xticks([i-0.4 for i in xs])
a.set_xticklabels(primerNames, rotation=90)
a.set_xlim(0, len(primerNames)+2)
a.set_ylabel('number of reads')
a.set_xlabel('primerID')
a.set_title('mappedReads_ham1_rl14_500k')
vizLib.cleanAxis(a)
figCounter = 3
pylab.tight_layout()
if saveFigs:
    pylab.savefig(figPath+str(figCounter)+'allBCsMappedReads.png', dpi=300)
    figCounter +=1
```
