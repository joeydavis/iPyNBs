```python
def findMax(column):    peaks = signal.find_peaks_cwt(column, numpy.arange(30,100,10))    maxPeaks = [column[i-20:i+20].argmax()+i-20 for i in peaks]    return maxPeaks
```

```python
def findMaxTimes(timeColumn, dataColumn):    index = findMax(dataColumn)    return [timeColumn[i] for i in index]
```

```python
def getInRange(peakTimes, range):    for i in peakTimes:        if i > range[0] and i < range[1]:            return i    return -1
```

```python
def peakInt(ys, xs, baseline, xl, xh):    sum = 0    for i in range(0, len(xs)):        ii = xs[i]        if ii >=xl and ii <= xh:            sum = sum + (ys[i]-baseline)*(xs[i+1] - xs[i])    return sum
```
