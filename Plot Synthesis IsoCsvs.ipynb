```python
num = ['AMP_L']
den = ['AMP_U']

msdPlus = qMS.multiStatsDict(plusIPTG, num, den, normalization=1.0, offset=0.0, normProtein=None)
msdNo = qMS.multiStatsDict(noIPTG, num, den, normalization=1.0, offset=0.0, normProtein=None)
```

```python
allListPlus = qMS.calculateMedian(msdPlus, plusIPTG, allSubunit)
for k in allListPlus.keys():
    allListPlus[k] = [i-.35 for i in allListPlus[k]]
    #allListPlus[k].insert(0, 0.0)
allListNo = qMS.calculateMedian(msdNo, noIPTG, allSubunit)
for k in allListNo.keys():
    allListNo[k] = [i-.35 for i in allListNo[k]]
    #allListNo[k].insert(0, 0.0)
```

```python
time = [0, 15, 30, 60, 100]
```

```python
sc2 = vizLib.proteinScatterPlot(allListNo, time, xMin = -5, xMax = 110, yMin = 0, yMax = 4, title='noIPTG', xLabel = 'time (mins)', yLabel = '50%15N/15N or doubling [Ai = 58], offset=1', figSize = (5.5,8), markerSize=100, linestyle='-')
t = numpy.linspace(0, 120, num=200)
sc2.plot(t, qMS.growthCurve(58, t, 1.0)-1, c='red', lw=5)
pylab.tight_layout()
pylab.savefig(figPath+'curveNoOffset.png')
```
