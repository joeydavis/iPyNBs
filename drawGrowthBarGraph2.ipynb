```python
import numpy
import scipy
from scipy import stats
import pylab
import csv
import os
```

```python
path = '/home/jhdavis/Dropbox/fromLinux'
files = [f for f in os.listdir(path)]
readFiles = {}
rowNames = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
colNames = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

allDat = {}
for i in rowNames:
    for j in colNames:
        allDat[i+j] = []
timePoints = [0, 47, 88, 134, 157, 260]
for f in files:
    fil = path+'/'+f
    readFiles[fil]=[]
    reader = csv.reader(open(fil, "rb"), delimiter='\t')
    for row, record in enumerate(reader):
        readFiles[fil].append(record)

ks = readFiles.keys()
ks.sort()
for k in ks:
    for r, row in enumerate(readFiles[k]):
        for c, col in enumerate(row):
            allDat[rowNames[r]+colNames[c]].append(col)

ks = allDat.keys()
ks.sort()

f = pylab.figure()
ax = f.add_subplot(111)
slope = {}
yVals = {}
for k in ks:
    yVals[k] = [numpy.log2(float(x)) for x in allDat[k]]
    s = float(stats.linregress(timePoints,yVals[k])[0])
    slope[k] = s
values = [float(x) for x in slope.values()]
ax.bar(range(96), values, 1)

ax.set_xticks(range(96))
ax.set_xticklabels( ks )
```
