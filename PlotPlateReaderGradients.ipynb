```python
path = '2013_04_16-NCMGradients/'
```

```python
files = ['blank.txt', 'plate1.txt', 'plate2.txt']header = [str(i) for i in range(1,13)]
```

```python
for i in files:    out = path+i[:-4]+'_crop.txt'    f = open(path+i)    of = open (out, 'w')    q = 8    while q > 0:        l = f.readline()        of.write(l)        q = q-1    f.close()    of.close()
```

```python
files_crop = ['blank_crop.txt', 'plate1_crop.txt', 'plate2_crop.txt']
```

```python
blank = pd.read_table(path+files_crop[0], names=header, index_col=False)
```

```python
plate1 = pd.read_table(path+files_crop[1], names=header, index_col=False)
```

```python
plate2 = pd.read_table(path+files_crop[2], names=header, index_col=False)
```

```python
plate1_corr = (((plate1-blank)-0.04)*10*1.5)/25.6
```

```python
plate2_corr = (((plate2-plate1)-0.04)*10*1.5)/25.6
```

```python
fracs = range(1,49)g1A = [plate1_corr[0:4].values[i] for i in range(0,4)]g1A = np.concatenate(g1A)g2A = [plate1_corr[4:8].values[i] for i in range(0,4)]g2A = np.concatenate(g2A)g3A = [plate2_corr[0:4].values[i] for i in range(0,4)]g3A = np.concatenate(g3A)g4A = [plate2_corr[4:8].values[i] for i in range(0,4)]g4A = np.concatenate(g4A)
```

```python
f = pylab.figure(figsize=(10,4))ax = f.add_subplot(111)ax.plot(fracs,g1A, label='wt(g1)', lw=2)ax.plot(fracs,g2A, label='L14(g2)', lw=2)ax.plot(fracs,g3A, label='L17(g3)', lw=2)ax.plot(fracs,g4A, label='L28(g4)', lw=2)pylab.xlabel('fractions')pylab.ylabel('pmol/uL 50S (ext = 25.6)')pylab.grid(b=True, which='major', linestyle='-')pylab.legend()pylab.tight_layout()pylab.savefig(figPath+'Full.png')pylab.savefig(figPath+'Full.pdf')
```
