```python
files = qMS.sort_nicely([i for i in glob.glob(path+'*.csv')])
labels = ['0:00', '2:40', '4:00', '8:00', '16:00']
DMSO = files[0:5]
CAM = files[5:10]
LAM = files[10:15]
ERM = files[15:20]
smallSubunit = qMSDefs.positionLabels30S[1:]
largeSubunit = qMSDefs.positionLabels50S[1:]
allSubunit = qMSDefs.positionLabels70S[1:]
allSubunit.remove('L12')
allSubunit.remove('S22')
allSubunit.remove('L8')
allSubunit.remove('L26')
```

```python
num = ['AMP_L']
den = ['AMP_S']

msdDMSO = qMS.multiStatsDict(DMSO, num, den, normalization=1.0, offset=0.0, normProtein=None)
msdCAM = qMS.multiStatsDict(CAM, num, den, normalization=1.0, offset=0.0, normProtein=None)
msdLAM = qMS.multiStatsDict(LAM, num, den, normalization=1.0, offset=0.0, normProtein=None)
msdERM = qMS.multiStatsDict(ERM, num, den, normalization=1.0, offset=0.0, normProtein=None)

allListDMSO = qMS.calculateMedian(msdDMSO, DMSO, allSubunit)
allListCAM = qMS.calculateMedian(msdCAM, CAM, allSubunit)
allListLAM = qMS.calculateMedian(msdLAM, LAM, allSubunit)
allListERM = qMS.calculateMedian(msdERM, ERM, allSubunit)
```

```python
time = [0, 160, 240, 480, 960]
```

```python
def getSlopes(l, t=[240, 480, 960]):
    slopeDict = {}
    for p in l.keys():
        v = l[p][2:]
        slopeDict[p] = max(numpy.polyfit(t, v, 1)[0],0.0)
    return slopeDict
```

```python
Blue = vizLib.getBCs('s', 5, name='Blues')[-2]
Red = vizLib.getBCs('s', 5, name='Reds')[-2]
Green = vizLib.getBCs('s', 5, name='Greens')[-2]
Purple = vizLib.getBCs('s', 5, name='Purples')[-2]
```

```python
slopesDMSO = getSlopes(allListDMSO)
slopesCAM = getSlopes(allListCAM)
slopesLAM = getSlopes(allListLAM)
slopesERM = getSlopes(allListERM)
```

```python
D = slopesDMSO
C = slopesCAM
L = slopesLAM
E = slopesERM
f = pylab.figure(figsize=(22,6))
a = f.add_subplot(111)
a.bar([i-0.4 for i in range(len(qMS.sort_nicely(D.keys())))], [D[i] for i in qMS.sort_nicely(D.keys())], align='edge', width=0.2, color=Blue, label='DMSO')
a.bar([i-0.2 for i in range(len(qMS.sort_nicely(C.keys())))], [C[i] for i in qMS.sort_nicely(C.keys())], align='edge', width=0.2, color=Red, label='CAM')
a.bar([i for i in range(len(qMS.sort_nicely(L.keys())))], [L[i] for i in qMS.sort_nicely(L.keys())], align='edge', width=0.2, color=Green, label='LAM')
a.bar([i + 0.2 for i in range(len(qMS.sort_nicely(E.keys())))], [E[i] for i in qMS.sort_nicely(E.keys())], align='edge', width=0.2, color=Purple, label='ERM')
pylab.legend()


a.set_xticks(range(54))
a.set_xlim(32,54)
a.set_ylim(0,0.0012)
a.set_yticks([0, 0.0004, 0.0008, 0.0012])
a.set_xticklabels(qMS.sort_nicely(D.keys()), size='medium', rotation='90')
a.set_ylabel('fit slope [using final 3 points]')
a.set_title('synthesis rate')
vizLib.cleanAxis(a,ticksOnly=True)
#pylab.savefig(savePath+'barGraphSynthesisRateSmall.png')
#pylab.savefig(savePath+'barGraphSynthesisRateSmall.pdf')
```
